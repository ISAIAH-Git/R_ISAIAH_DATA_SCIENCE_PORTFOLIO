```python
url='https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page=requests.get(url)
soup=BeautifulSoup(page.text,'html.parser')
soup
```

```python
countries=soup.find_all(class_='flagicon')
countries[0]
```

```python
rank=[]
country=[]
link=[]
population=[]
world=[]
date=[]
source=[]
```

```python
for i in range(1,21):
    table=soup.find_all('tr')
    rank.append(table[i].find('th').text.strip('\n'))
    country.append(table[i].find('a').text)
    link.append(table[i].find('a')['href'])
    z=table[i].find_all('td')
    population.append(z[1].text)
    world.append(z[2].text)
    date.append(z[3].text)
    source.append(z[4].text.strip('\n'))
```

```python
#Lets Create DataFrame
```

```python
wikipedia=pd.DataFrame({'Rank':rank,'Country':country,'link':link,'Population':population,'Percentage of World':world,'Date':date,'source':source})
```
