```python
import numpy as np
import matplotlib.pyplot as plt
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)
sinplot(-1)
```

```python
import seaborn as sns
sns.set_style('whitegrid')
sinplot()
```

```python
sns.set_style('dark')
sinplot()
```

```python
sns.set_style('white')
sinplot()
```

```python
sns.set_style('ticks')
sinplot()
```

```python
sns.set_style('ticks')
sinplot()
sns.despine()
```

```python
sns.set_style('ticks')
sinplot()
sns.despine(left=True,bottom=True)
```

```python
sns.set_style('ticks',{'axes.grid':True,'xtick.direction':'in'})
sinplot()
sns.despine(left=True,bottom=False)
```
