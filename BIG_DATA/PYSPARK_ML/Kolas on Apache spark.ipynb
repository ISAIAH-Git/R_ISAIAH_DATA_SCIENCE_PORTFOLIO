```
import pandas as pd
import databricks.koalas as ks
```

```
loan_data=ks.read_csv("dbfs:/FileStore/tables/LoanStats_2018Q4-1.csv")
loan_data.head() #Data should be in local file system rather than distributed to run pandas.
```

```
loan_data["loan_status"].value_counts()
```

```
ks.set_option("display.max_rows",50)
```

```
ld_sel=loan_data[['loan_amnt','grade','term','loan_status']]
```

```
ld_sel['loan_status'].value_counts()
```

```
ld_sel.groupby("grade").agg({"grade":"count"})
```

```
ld_sel.groupby("grade").agg({"grade":"count"})/ld_sel["grade"].count()
```

```
ld_tmp=ks.concat([ld_sel.groupby("grade").agg(grade_count=("grade","count")),ld_sel.groupby("grade").agg(grade_perct=("grade","count"))/ld_sel['grade'].count()],axis=1)
```

```
ld_tmp.head()
```

```
ld_tmp.reset_index() #Here spark has no indexing but kolas has indexing as it replicates pandas
```

```
ks.sql("SELECT * FROM {ld_tmp} WHERE grade_count>300")
```

```
ld_tmp.sort_index()
```

```
ld_tmp.sort_values(by="grade_perct")
```

```
import numpy as np
ld_tmp.apply(np.sum)
```

```
ld_tmp=ld_tmp.dropna()
```

```
display(ld_tmp)
```

```
ld_tmp.apply(lambda x:x*2)
```

```
def multiply(x) ->ks.Series[np.float64]:
  return x*2
```

```
ld_tmp.apply(multiply)
```
