```
#TRANSFORMATION TYPES:
# 1) NARROW TRANFORMATION  : ALL ELEMENTS THAT ARE REQUIRED TO COMPUTE THE RECORD IS IN A SINGLE PARTITATION
# I.E EACH PARTITION OF THE PARENT RDD IS USED BY AT MOST ONE PARTITION OF THE CHILD RDD LIKE : MAP , FILTER, UNION , CO PARTITIONED


# 2) WIDE TRANSFORMATION : ALL ELEMENTS THAT ARE REQUIRED TO COMPUTE THE RECORD IS IN A MUTILPLE PARTITIONS
#I.E MULTIPLE CHILD RDD PARTITIONS MAY DEPEND ON A SINGLE PARENT RDD PARTITION LIKE GROUP BY KEY INPUTS NOT CO PARTITIONED 

#LOT OF DATA MOVING ACROSS DIFFERENT SYSTEMS.
```

```
#SOMETIMES YOU MIGHT NEED RDD WHENEVER UR DOING A DATA FRAME FUNCTION GOING TO AN UNDERLINYING LOW LEVEL API MIGHT GIVE YOU A BETTER FELEXIBILITY.
```

```
#RDD'S ARE OUTDATED DATAFRAMES ARE THE SYNTAX WE SHOULD BE USING.
```

```
x=spark.sparkContext.parallelize([1,4,8])
y=x.flatMap(lambda x:(x,x*x))
print(x.collect())
print(y.collect())
```

```
x=spark.sparkContext.parallelize([1,4,8])
y=x.map(lambda x:(x,x*x))
print(x.collect())
print(y.collect())
```

```
import pyspark.sql.functions as F
corona_df.groupBy("Country","State_cleaned").agg(F.max("Date")).show()
```

```
corona_df.join(corona_df.groupBy("Country","State_cleaned").agg(F.max("Date").alias("Date")),on=["Country","State_cleaned","Date"],how="inner")
```

```
corona_df.join(corona_df.groupBy("Country","State_cleaned").agg(F.max("Date").alias("Date")),on=['Country', 'State_cleaned','Date'],how="inner").sort("Confirmed",ascending=False).show(10)
```

```
from pyspark.sql.functions import row_number
from pyspark.sql.window import Window

ws=Window().partitionBy("Country","State_cleaned").orderBy(col("Date").desc())

corona_df.withColumn("row_num", row_number().over(ws)).show()
```

```
corona_df.withColumn("row_num", row_number().over(ws)).where(col("row_num")==1).show()
```

```
corona_max_df=corona_df.join(corona_df.groupBy("Country","State_cleaned").agg(F.max("Date").alias("Date")),on=['Country', 'State_cleaned','Date'],how="inner")
```
